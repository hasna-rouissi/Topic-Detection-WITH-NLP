```python
# Objectif : √âvaluer le mod√®le de d√©tection de nouveaut√©


import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import (
    accuracy_score, precision_score, recall_score, f1_score,
    confusion_matrix, classification_report, log_loss
)
import matplotlib.pyplot as plt
import seaborn as sns
import os
from tqdm import tqdm

# === 1. Chargement des donn√©es ===
DATA_PATH = r"C:\Users\hasna\Desktop\Master\S3\NLP\Projet-NLP\data"

print(" Chargement du dataset enrichi avec labels...")
df = pd.read_csv(os.path.join(DATA_PATH, "data_small_with_labels.csv"))
embeddings = np.load(os.path.join(DATA_PATH, "embeddings_small_minilm.npy"))

print(f" Dataset charg√© : {df.shape}")
print(df.head())

# === 2. Split historique / test ===

#On simule une chronologie r√©elle : les 70 % premiers articles repr√©sentent l‚Äôhistorique connu
history_size = int(0.7 * len(df))
X_history = embeddings[:history_size] 
y_history = df["label"][:history_size].values


#Les 30 % restants repr√©sentent les nouveaux articles √† tester
X_test = embeddings[history_size:]
y_test = df["label"][history_size:].values


print(" \n Cela permet d‚Äô√©valuer le mod√®le dans un contexte r√©aliste de flux d‚Äôarticles arrivant dans le temps.")
print(f"\n Historique : {len(X_history)} textes")
print(f" Test : {len(X_test)} textes")



# === 3. D√©tection de la nouveaut√© sur le test ===

threshold = 0.60  # seuil de similarit√© pour d√©cider si c'est nouveau
pred_labels = []

print("\n D√©tection de la nouveaut√© sur le test...")
for i in tqdm(range(len(X_test))):
    # Calcul de la similarit√© cosinus avec l'historique
    
    sims = np.dot(X_history, X_test[i]) #Pour chaque embedding du test ,on calcule sa similarit√© cosinus avec tous les embeddings de l‚Äôhistorique
    max_sim = np.max(sims) 
    
    # Label bas√© sur le seuil
    if max_sim < threshold:
        pred_labels.append(1)  # nouveau
    else:
        pred_labels.append(0)  # similaire

pred_labels = np.array(pred_labels)

# === 4.  Calcul des m√©triques classiques ===

f1 = f1_score(y_test, pred_labels)  #√©quilibre entre pr√©cision et rappel
prec = precision_score(y_test, pred_labels) #fiabilit√© des pr√©dictions ‚Äúnouveau‚Äù
rec = recall_score(y_test, pred_labels)  #capacit√© √† d√©tecter tous les vrais ‚Äúnouveaux‚Äù
acc = accuracy_score(y_test, pred_labels)   #pourcentage de pr√©dictions correctes

print("\n === R√©sultats d'√©valuation ===")
print(f"F1-Score  : {f1:.4f}")
print(f"Precision : {prec:.4f}")
print(f"Recall    : {rec:.4f}")
print(f"Accuracy  : {acc:.4f}")

print("\n=== Rapport complet ===")
print(classification_report(y_test, pred_labels))

# === 5. Binary Cross-Entropy Loss ===
# Probabilit√© que le texte soit nouveau : 1 - max_sim
prob_new = []
for i in range(len(X_test)):
    sims = np.dot(X_history, X_test[i])
    max_sim = np.max(sims)
    prob_new.append(1 - max_sim)

prob_new = np.clip(prob_new, 1e-7, 1-1e-7)  # √©viter log(0)
loss = log_loss(y_test, prob_new)
print(f"\n Binary Cross-Entropy Loss : {loss:.4f}")

# === 6. Matrice de confusion ===
plt.figure(figsize=(5,4))
sns.heatmap(confusion_matrix(y_test, pred_labels), annot=True, fmt='d', cmap='Blues')
plt.title("Matrice de confusion")
plt.xlabel("Pr√©diction")
plt.ylabel("V√©rit√© terrain")
plt.show()

# === 7. Analyse de sensibilit√© du seuil ===
thresholds = np.linspace(0.4, 0.8, 9)
f1_scores = []

for th in thresholds:
    pred_th = []
    for i in range(len(X_test)):
        sims = np.dot(X_history, X_test[i])
        max_sim = np.max(sims)
        pred_th.append(1 if max_sim < th else 0)
    f1_scores.append(f1_score(y_test, pred_th))

plt.figure(figsize=(6,4))
plt.plot(thresholds, f1_scores, marker='o')
plt.title("Sensibilit√© du F1-score au seuil de similarit√©")
plt.xlabel("Seuil de similarit√©")
plt.ylabel("F1-Score")
plt.grid(True)
plt.show()
```

```python
##Interpr√©tation de ton score : 0.56

#Une loss de 0.56 indique que ton mod√®le fait des pr√©dictions coh√©rentes, mais pas encore parfaitement calibr√©es.
#Il distingue correctement les textes nouveaux des anciens (tu avais un F1 ‚âà 0.95 !),
#mais ses probabilit√©s associ√©es ne sont pas toujours tr√®s bien ajust√©es.

#En d‚Äôautres termes : il classe bien, mais n‚Äôest pas encore parfaitement s√ªr de ses degr√©s de nouveaut√©.
```
