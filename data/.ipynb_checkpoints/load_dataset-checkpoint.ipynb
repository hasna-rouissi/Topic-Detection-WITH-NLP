```python
import pandas as pd

# === Charger le fichier brut ===
file_path = "data with abstract"

# Tentatives avec diff√©rents s√©parateurs
try:
    df = pd.read_csv(file_path, sep=",")
except Exception:
    try:
        df = pd.read_csv(file_path, sep=";")
    except Exception:
        df = pd.read_csv(file_path, sep="\t")

# ===  V√©rifier les colonnes ===
print("Colonnes d√©tect√©es :", df.columns)

# Si toutes les colonnes sont dans une seule
if len(df.columns) == 1:
    print(" Donn√©es non s√©par√©es ‚Äî tentative de correction...")
    # Forcer la s√©paration manuelle
    df = df.iloc[:, 0].str.split(",", expand=True)
    df.columns = ["PY", "id", "eid", "TI", "author"]

# ===  Nettoyer les valeurs ===
df = df.dropna(subset=["TI"])  # Supprimer les lignes sans titre
df = df.reset_index(drop=True)

print(" Aper√ßu du dataset :")


# === Sauvegarder en CSV propre ===
df.to_csv("data_clean.csv", index=False)
print(" Fichier corrig√© enregistr√© sous data_clean.csv")
```

```python
import pandas as pd


# Calcul du nombre de mots pour chaque texte
df["word_count_TI"] = df["TI"].astype(str).apply(lambda x: len(x.split()))

# Affichage des premi√®res lignes
print(df[["TI", "word_count_TI"]].head())

# Calcul des intervalles
min_words = df["word_count_TI"].min()
max_words = df["word_count_TI"].max()
mean_words = df["word_count_TI"].mean()

print("\n R√©sultats :")
print(f" Nombre minimum de mots : {min_words}")
print(f" Nombre maximum de mots : {max_words}")
print(f" Moyenne de mots : {mean_words:.2f}")
```
